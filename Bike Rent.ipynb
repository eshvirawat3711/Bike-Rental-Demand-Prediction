# Statistical Feature Selection (Categorical Vs Continuous) using ANOVA test
Analysis of variance(ANOVA) is performed to check if there is any relationship between the given continuous and categorical variable
* Assumption(H0): There is NO relation between the given variables (i.e. The average(mean) values of the numeric Target variable is same for all the groups in the categorical Predictor variable)
* ANOVA Test result: Probability of H0 being true

```python
from scipy.stats import f_oneway
```

```python
# Defining a function to find the statistical relationship with all the categorical variables
def AnovaFunc(InpData,TargetVariable,CatPreds):
    
     # Creating an empty list of final selected predictors
    SelectedPreds=[]
    
    for pred in CatPreds:
        CatPredList=InpData.groupby(pred)[TargetVariable].apply(list)
        Anova=f_oneway(*CatPredList) #unpacks the list
    
        # If the ANOVA P-Value is <0.05, that means we reject H0
        if(Anova[1]<0.05):
            print(pred,'is correlated with', TargetVariable, '| P-Value:', Anova[1])
            SelectedPreds.append(predictor)
        else:
            # Accepting the H0 if the P value is more than 0.05
            print(predictor, 'is NOT correlated with', TargetVariable, '| P-Value:', Anova[1])
    
    return(SelectedPreds)
        
```

```python
AnovaFunc(InpData=BikesRentData,TargetVariable='cnt',CatPreds=CatCols)
```

All categorical variables are correlated with the Target variable.

<b>Final selected Categorical columns:</b>
['season', 'mnth', 'hr', 'holiday', 'weekday', 'workingday', 'weathersit']

# Machine Learning: Splitting the data into Training and Testing sample

```python
x=final_df[Predictors]
y=final_df['cnt']
```

```python
#splitting the data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

# Multiple Linear Regression

```python
from sklearn.linear_model import LinearRegression
lm=LinearRegression()
lm.fit(x_train,y_train)
```

# Model Evaluation

```python
from sklearn import metrics

# Measuring Goodness of fit in Training data
print('R2 Value:',metrics.r2_score(y_train,lm.predict(x_train)))
```

```python
prediction=lm.predict(x_test)
plt.scatter(prediction,y_test)
```

```python
mape=np.mean(abs((y_test-prediction)/y_test))*100
med=np.median(abs((y_test-prediction)/y_test))*100

Accuracy=100-mape
MedAccuracy=100-med

print('mean accuracy on test data', Accuracy)
print('median accuracy on test data',MedAccuracy)
```

```python
# Defining a custom function to calculate accuracy
def Accuracy_Score(orig,pred):
    mape=np.mean(abs((orig-pred)/orig))*100
    return(100-mape)

# Custom Scoring MAPE calculation
from sklearn.metrics import make_scorer
cust_score=make_scorer(Accuracy_Score)

# Importing cross validation function from sklearn
from sklearn.model_selection import cross_val_score
# Passing full data X and y because the K-fold will split the data and automatically choose train/test
print('Avg model accuracy:',round((cross_val_score(lm,x,y,cv=5,scoring=cust_score)).mean(),2))
```

# Decision Trees

```python
from sklearn.tree import DecisionTreeRegressor
rm=DecisionTreeRegressor(max_depth=3,criterion='mse')
rm.fit(x_train,y_train)
```

```python
# Measuring Goodness of fit in Training data
print('R2 Value:',metrics.r2_score(y_train,rm.predict(x_train)))
```

```python
# Plotting the feature importance for Top 10 most important columns
pd.Series(rm.feature_importances_,index=Predictors).nlargest(5).plot(kind='barh')
```

```python
prediction=rm.predict(x_test)
mape=np.mean(abs((y_test-prediction)/y_test))*100
med=np.median(abs((y_test-prediction)/y_test))*100

print('Mean Accuracy',100-mape)
print('Median Accuracy',100-med)
print('Avg model Accuracy',round((cross_val_score(rm,x,y,cv=5,scoring=cust_score)).mean(),2))
```
